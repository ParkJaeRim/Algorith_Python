```
from collections import deque

q=deque()
case = int(input())

dx = [1,-1,0,0]
dy = [0,0,1,-1]

for i in range(case):
  N = int(input())
  path = [[0]*N for _ in range(N)]
  visited = [[False]*N for _ in range(N)]
  cnt = [[-1]*N for _ in range(N)]
  for j in range(N):
    path[j] = list(map(int, input()))

  for x in range(N):
    for y in range(N):
      if path[x][y] == 2:
        q.append((x,y))
        visited[x][y] = True
        cnt[x][y] = 0
      if path[x][y] == 3:
        end_x = x
        end_y = y

  while q:
    x,y = q.popleft()
    for k in range(4):
      nx, ny = x+dx[k] , y+dy[k]

      if 0 <= nx < N and 0 <= ny < N :
        if path[nx][ny] == 0 and visited[nx][ny] == False:
          visited[nx][ny] = True
          cnt[nx][ny] = cnt[x][y] +1
          q.append((nx,ny))
        if path[nx][ny] == 3 and visited[nx][ny] == False:
          visited[nx][ny] = True
          cnt[nx][ny] = cnt[x][y] +1

  if visited[end_x][end_y] == True:
    print('#{0} {1}'.format(i+1, cnt[end_x][end_y]-1))
  else:
    print('#{0} {1}'.format(i+1, 0))
```
